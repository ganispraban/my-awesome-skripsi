```python
import matplotlib.pyplot as plt
%matplotlib inline

df0 = dataset[:30]
df1 = dataset[30:60]
df2 = dataset[60:90]
df3 = dataset[90:120]
df4 = dataset[120:150]

plt.figure(figsize=(15,7))
plt.scatter(df0['contrast'], df0['homogeneity'], df0['ASM'], df0['energy'], marker='.',label='Skala0')
plt.scatter(df1['contrast'], df1['homogeneity'], df1['ASM'], df1['energy'], marker=',',label='Skala1')
plt.scatter(df2['contrast'], df2['homogeneity'], df2['ASM'], df2['energy'], marker='o',label='Skala2')
plt.scatter(df3['contrast'], df3['homogeneity'], df3['ASM'], df3['energy'], marker='v',label='Skala3')
plt.scatter(df4['contrast'], df4['homogeneity'], df4['ASM'], df4['energy'], marker='p',label='Skala4')
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X = dataset.drop(['target', 'class'], axis='columns')
y = dataset.target

X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.2, random_state=3)
```

```python
X_train.shape
```

```python
y_train.shape
```

#### Create K Nearest Neighbor Classification (KNN)

```python
from sklearn.neighbors import KNeighborsClassifier
knn= KNeighborsClassifier(n_neighbors=11)
knn.fit(X_train,y_train)
```

```python
knn.score(X_test,y_test)
```

#### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
y_pred=knn.predict(X_test)
matrix= confusion_matrix(y_test, y_pred)  #  confusion_matrix(truth, prediction)
matrix
```

```python
import seaborn as sns
plt.figure(figsize=(10,8))
sns.heatmap(matrix, annot= True)
plt.xlabel('predict')
plt.ylabel('truth')
```

#### Classification report

```python
from sklearn.metrics import classification_report
print( classification_report(y_test,y_pred))
```
