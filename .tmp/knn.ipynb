```python
import pandas as pd
import math
import csv
from sklearn import preprocessing
import random
import numpy as np

from sklearn.metrics import confusion_matrix 
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report
```

```python
def knn(k, training_data, testing_data, training_class, test_class, actual=[], predicted=[]):
    for test_instance, tes_instance in zip(testing_data, test_class):
        row3=[]
        row2=[]
        row1=[]
        for training_instance,train_instance in zip(training_data,training_class):
            # print(training_instance[0])
            # dist = abs(float(test_instance[0]) - float(training_instance[0]))
            distance=pow((float(test_instance[0]) - float(training_instance[0])), 2)
            dist=math.sqrt(distance)
            row_dict={'predicted_class':train_instance,'actual_class':tes_instance,'value':dist}
            row1.append(row_dict)
        #print(row1)
        row2=sorted(row1, key = lambda i: i['value'])
        #print(len(row2))
        
        for i in range(k):
            row3.append(row2[i])
        #test_data.append(row3)
    
        test_array=np.asarray(row3)
        
        #for i in test_array:
            #print(i)
        skala0_count=0
        skala4_count=0
        for j in test_array:
            if(j.get('predicted_class')=='1'):
                skala0_count+=1
            else:
                skala4_count+=1

        if(skala0_count >= skala4_count):
            predicted.append('1')
            actual.append(j.get('actual_class'))
        else:
            row3=[]
            row2=[]
            row1=[]
            
            for training_instance, train_instance in zip(training_data,training_class):
                distance=0
                for x in range(1,8):
                    distance += pow(float(test_instance[x]) - float(training_instance[x]), 2)
                dist=math.sqrt(distance)
                row_dict={'predicted_class':train_instance,'actual_class':tes_instance,'value':dist}
                row1.append(row_dict)
            
            row2=sorted(row1, key = lambda i: i['value'])
               
            for i in range(k):
                row3.append(row2[i])

            test_array=np.asarray(row3)
                
            skala1_count=0
            skala2_count=0
            skala3_count=0
            # skala0_count=0
            for j in test_array:
                if(j.get('predicted_class')=='0'):
                    skala1_count+=1
                elif(j.get('predicted_class')=='1'):
                    skala2_count+=1
                elif(j.get('predicted_class')=='2'):
                     skala3_count+=1
                elif(j.get('predicted_class')=='3'):
                    skala3_count+=1

            if(skala1_count >= skala2_count):
                predicted.append('1')
                actual.append(j.get('actual_class'))
            elif(skala2_count>= skala3_count):
                predicted.append('2')
                actual.append(j.get('actual_class'))
            elif(skala3_count>= skala4_count):
                predicted.append('3')
                actual.append(j.get('actual_class'))
            else:
                predicted.append('4')
                actual.append(j.get('actual_class'))
    print(actual, predicted)
```

```python

def loadDataset(filename, split, trainingSet=[] , testSet=[], trainingClass=[],testClass=[]):
    arr=[]
    names = ['sumValue','contrast', 'homogeneity', 'ASM', 'energy','hue', 'saturaton', 'value','label']
    dataset = pd.read_csv(filename, names=names)
    
    X = dataset.iloc[:, :-1].values
    y = dataset.iloc[:, 8].values

    print('length of dataset', len(dataset))
    
    for x in range(1,len(dataset)):
        if random.random() <= split:
            trainingSet.append(X[x])
            trainingClass.append(y[x])
        else:
            testSet.append(X[x])
            testClass.append(y[x])
    
```
